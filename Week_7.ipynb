```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random, seed
```

# **Exercise 1**


1.)

```python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    '''
    x = np.zeros(steps, int)  # initialise the array of integers
    x[0] = 0  # start at the origin
    for i in range(1, len(x)):
        rnd = random()  # rnd selected from [0,1)
        # step left or right randomly ...
        if (rnd<0.5):
            x[i] = x[i-1] + 1  # steps right one unit
        else:
            x[i] = x[i-1] - 1 # steps left one unit
    return x

num_steps = 1000 #Plot for a 1000 steps
positions = walk(num_steps)
step_numbers = np.arange(num_steps)

plt.figure(figsize=(10, 6))
plt.plot(step_numbers, positions)
plt.title('1D Random Walk (1000 Steps)')
plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.grid(True)
```

```python
num_steps = 1000
num_walks = 100

step_numbers = np.arange(num_steps)

plt.figure(figsize=(10, 6)) # Create the plot

for _ in range(num_walks): # Loop to generate and plot 100 walks
    positions = walk(num_steps)
    plt.plot(step_numbers, positions)

plt.title('100 Random Walks (1000 Steps Each)')
plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.grid(True)
```

```python
num_steps = 1000
N = 10  # Number of walkers

# Array shape is (steps, N)
particles = np.zeros((num_steps, N), int)
# All walkers start at origin

# Loop through each step from 1 onwards
for i in range(1, num_steps):
    # Loop through each particle (walker)
    for j in range(N):
        rnd = random()  # Get a random number

        # Decide step direction based on the previous step of the *same particle*
        if (rnd < 0.5):
            particles[i, j] = particles[i-1, j] + 1  # Step right
        else:
            particles[i, j] = particles[i-1, j] - 1  # Step left

step_numbers = np.arange(num_steps)

plt.figure(figsize=(10, 6))

# Loop through each particle (column) and plot its walk
for j in range(N):
    plt.plot(step_numbers, particles[:, j]) # Plot the j-th column

plt.title(f'{N} Random Walks (1000 Steps Each)')
plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.grid(True)
```

```python
num_steps = 1000
N = 500

# Array shape is (steps, N)
particles = np.zeros((num_steps, N), int)
# All walkers start at origin (row 0 is already all zeros)

# Loop through each step from 1 onwards
for i in range(1, num_steps):
    # Loop through each particle (walker)
    for j in range(N):
        rnd = random()  # Get a random number

        # Decide step direction based on the previous step of the *same particle*
        if (rnd < 0.5):
            particles[i, j] = particles[i-1, j] + 1
        else:
            particles[i, j] = particles[i-1, j] - 1

# np.mean(axis=1) calculates the mean across the rows (i.e., for all particles at each step)
mean_pos = np.mean(particles, axis=1)
mean_sq_pos = np.mean(np.square(particles), axis=1)

last_step_index = num_steps - 1
print(f"Mean Position: {mean_pos[last_step_index]}")
print(f"Mean Squared Position: {mean_sq_pos[last_step_index]}")

step_numbers = np.arange(num_steps)

plt.figure(figsize=(10, 6))

# Loop through each particle (column) and plot its walk
for j in range(N):
    # Use alpha for better visibility with many lines
    plt.plot(step_numbers, particles[:, j], alpha=0.1)

plt.title(f'{N} Random Walks (1000 Steps Each)')
plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.grid(True)
```

```python
mean_pos = np.mean(particles, axis=1)
step_numbers = np.arange(num_steps)

plt.figure(figsize=(10, 6))
plt.plot(step_numbers, mean_pos)
plt.title(f'Mean Position of {N} Walkers vs. Step Number')
plt.xlabel('Step Number (i)')
plt.ylabel('Mean Position <x>')
plt.grid(True)
# Add a horizontal line at y=0 for reference
plt.axhline(0, color='red', linestyle='--', linewidth=0.8)
```

```python
mean_sq_pos = np.mean(np.square(particles), axis=1)
step_numbers = np.arange(num_steps)

plt.figure(figsize=(10, 6))
plt.plot(step_numbers, mean_sq_pos, label='Mean Squared Position $\langle x^2 \\rangle$')
plt.title(f'Mean Squared Position of {N} Walkers vs. Step Number')
plt.xlabel('Step Number (i)')
plt.ylabel('Mean Squared Position $\langle x^2 \\rangle$')
plt.legend()
plt.grid(True)
```
